```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import urllib.parse
from fake_useragent import UserAgent
from time import sleep
import random
from tqdm import tqdm
```

```python
import numpy as np
```

```python
ua = UserAgent()
user_agent = {'User-agent': ua.random}
main_link = "https://www.pharm24.gr/bitamines-c-5083.html" 
user_agent = {'User-agent': ua.random}
response = requests.get(main_link, timeout=15, headers=user_agent)
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
pagination = soup.find('div', class_='pagination')
if pagination:
    max_pages_element = pagination.find_all('a')[-2]  
    max_pages = int(max_pages_element.text.strip())
    print(f"Max Pages: {max_pages}")
else:
    print("Max pages element not found.")
```

```python
#lets iterate over the range of pages and save all the links in a list
links = []

for page in range(1,int(max_pages)+1):
    link = main_link + '?page=' + str(page) + '/'
    links.append(link)
```

```python
#print 10
links[0:10]
```

```python
product_links =[]
```

```python
for link in links:
    response = requests.get(link, timeout=15, headers=user_agent, verify=False)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        cards = soup.find_all('div', class_='pWrap grid-style')
        for card in cards:
            product_link = 'https://www.pharm24.gr' + card.find('a').get('href')
            product_links.append(product_link)
```

```python
len(product_links)
```

```python
categories = []

for plink in tqdm(product_links[1:2]):
    response = requests.get(plink, timeout=15, headers=user_agent, verify=False)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        category_div = soup.find('div', id='see-also')
        if category_div: 
            category_links = category_div.find_all('a')
            category_text = [link.text for link in category_links] 
            categories.extend(category_text)  
print(categories)
```

```python
columns = ['product_name','category','normal_price','discount_price','reviews_count','rating']
df = pd.DataFrame(columns=columns)
df
```

```python
for plink in tqdm(product_links):
    response = requests.get(plink, timeout=15, headers=user_agent, verify=False)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        try:
            product_name = soup.find('h1', class_='no-margin').text
        except:
            product_name = np.nan
        try:
            category_div = soup.find('div', id='see-also')
            if category_div:
                category_links = category_div.find_all('a')
                category_text = [link.text for link in category_links]
                category = category_text
            else:
                category = np.nan
        except:
            category = np.nan        
        try:
            normal_price = soup.find('span', class_='normalprice').text
        except:
            normal_price = np.nan
        try:
            discount_price = soup.find('span', class_='productSalePrice').text.strip()
        except:
            discount_price = np.nan
        try:
            reviews_count = soup.find('div', class_='tt-p-reviews').text
        except:
            reviews_count = np.nan
        try:
            rating = soup.find('span', class_='tt-rating-num').text
        except:
            rating = np.nan

        temp = {
            'product_name': product_name,
            'category': category,
            'normal_price': normal_price,
            'discount_price': discount_price,
            'reviews_count': reviews_count,
            'rating': rating  
            }
        df = df.append(temp, ignore_index=True)
        sleep(0.6)
    else:
        print('Server is down!')
```
